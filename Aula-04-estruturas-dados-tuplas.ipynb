1. Considere a seguinte tupla: `(5,6,1,2,4,5,2,1,3,1)`. Desenvolva um ciclo `while` para visitar (iterar) cada objeto da coleção do primeiro até o último. Imprima o objeto a cada ciclo.

````python
num = (5,6,1,2,4,5,2,1,3,1) 
tamanho = len(num)
indice = 0

while indice < tamanho:
    print(num[indice])
    indice += 1
    
````

2. Repita o exercício da questão (1), todavia seguindo a ordem inversa (do último até o primeiro objeto). Imprima o objeto a cada ciclo.

````python
num = (5,6,1,2,4,5,2,1,3,1) 
tamanho = len(num)
indice = tamanho  -1

while indice >= 0 :
    print(num[indice])
    indice -= 1
````

3. Considere a tupla da questão (1). Faça um algoritmo que visite cada objeto, vefique números ímpares e pares e os armazene em listas separadas `pares` e `impares`. No final, imprima as listas.

````python
numeros = (5,6,1,2,4,5,2,1,3,1) 

pares = []
impares = []

tamanho = len(numeros)
indice = 0

while indice < tamanho:
    num = (numeros[indice])
    if num % 2 == 0:
        pares.append(num)
    else :
        impares.append(num)
indice += 1

print("Números pares:", pares)
print("Números ímpares:", impares)
````

5. Elabore um algoritmo que receba a temperatura média de cada mês do ano e armazene-as em uma lista. Também armazene os nomes de cada mês em outra lista. Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual e em que mês
elas ocorreram.

````python
nomes_meses = []
temperaturas = []

total_meses = 12


soma_temperaturas = 0


contador_meses = 1


while contador_meses <= total_meses:
    
    mes = input(f"Digite o nome do mês {contador_meses}: ")
    temperatura = float(input(f"Digite a temperatura média de {mes}: "))
    
    
    nomes_meses.append(mes)
    temperaturas.append(temperatura)
    
    
    soma_temperaturas += temperatura
    
    
    contador_meses += 1


media_anual = soma_temperaturas / total_meses


print("Média anual das temperaturas:", media_anual)


print("Temperaturas acima da média anual:")
indice = 0
while indice < total_meses:
    if temperaturas[indice] > media_anual:
        print(f"{nomes_meses[indice]}: {temperaturas[indice]}")
    indice += 1
````

6. Utilizando listas, faça um algoritmo com 5 perguntas para uma pessoa sobre um crime. As perguntas são:
   - "Telefonou para a vítima?"
   - "Esteve no local do crime?"
   - "Mora perto da vítima?"
   - "Devia para a vítima?"
   - "Já trabalhou com a vítima?"

O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".

````python
respostas = []


perguntas = [
    "Telefonou para a vítima?",
    "Esteve no local do crime?",
    "Mora perto da vítima?",
    "Devia para a vítima?",
    "Já trabalhou com a vítima?"
]


contador_perguntas = 0


while contador_perguntas < len(perguntas):
    resposta = input(perguntas[contador_perguntas] + " (sim/não): ")
    if resposta.lower() == "sim" or resposta.lower() == "s":
        respostas.append(1)  
    else:
        respostas.append(0)  
    contador_perguntas += 1


soma_respostas = sum(respostas)


if soma_respostas == 5:
    print("Assassino")
elif soma_respostas >= 3:
    print("Cúmplice")
elif soma_respostas >= 2:
    print("Suspeita")
else:
    print("Inocente")
````

7. Faça um algorimo que leia um número indeterminado de valores, correspondentes a notas entre 0 e 10 sobre avaliação de um produto. Encerre a entrada de dados quando for informado um valor igual a -1 (que não deve ser armazenado). Após esta entrada de dados, faça o seguinte:

   - Mostre a quantidade de valores que foram lidos;
   - Exiba todos os valores na ordem em que foram informados, um ao lado do outro;
   - Exiba todos os valores na ordem inversa à que foram informados, um abaixo do outro;
   - Calcule e mostre a soma dos valores;
   - Calcule e mostre a média dos valores;
   - Calcule e mostre a quantidade de valores acima da média calculada;
   - Calcule e mostre a quantidade de valores abaixo da média;
   - Encerre o programa com uma mensagem condicionada: o produto será classificado com avaliação positiva e a média das novas for superior à 7.0, e negativa caso contrário.

````python
notas = []


while True:
    nota = float(input("Digite a nota (ou -1 para encerrar): "))
    if nota == -1:
        break  
    if nota < 0 or nota > 10:
        print("Nota inválida. Por favor, digite uma nota entre 0 e 10.")
        continue  
    notas.append(nota)


print("Quantidade de valores lidos:", len(notas))


print("Valores na ordem em que foram informados:", *notas)


print("Valores na ordem inversa à que foram informados:")
for nota in reversed(notas):
    print(nota)


soma = sum(notas)
print("Soma dos valores:", soma)


media = soma / len(notas)
print("Média dos valores:", media)


acima_da_media = sum(1 for nota in notas if nota > media)
abaixo_da_media = sum(1 for nota in notas if nota < media)
print("Quantidade de valores acima da média:", acima_da_media)
print("Quantidade de valores abaixo da média:", abaixo_da_media)


if media > 7.0:
    print("O produto será classificado com avaliação positiva.")
else:
    print("O produto será classificado com avaliação negativa.")
````

8. Uma grande emissora de televisão quer fazer uma enquete entre os seus telespectadores para saber qual o melhor jogador após cada jogo de futebol. Para isto, faz-se necessário o desenvolvimento de um programa, que será utilizado pelas telefonistas, para a computação dos votos. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação Python. Para computar cada voto, a telefonista digitará um número, entre 1 e 23,
correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada. Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:
   - O total de votos computados;
   - Os números e respectivos votos de todos os jogadores que receberam votos;
   - O número do jogador escolhido como o melhor jogador da partida.

Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer **uso de listas**. 
Exemplo de saída esperada:
```text
   Enquete: Quem foi o melhor jogador?

   Número do jogador (0=fim): 9
   Número do jogador (0=fim): 10
   Número do jogador (0=fim): 9
   Número do jogador (0=fim): 10
   Número do jogador (0=fim): 11
   Número do jogador (0=fim): 10
   Número do jogador (0=fim): 50
   Informe um valor entre 1 e 23 ou 0 para sair!
   Número do jogador (0=fim): 9
   Número do jogador (0=fim): 9
   Número do jogador (0=fim): 0

   Resultado da votação:

   Foram computados 8 votos.

   Jogador Votos           %
   9               4              
   10              3               
   11              1               
   O melhor jogador foi o número 9, com 4 votos de um total de 7.
```

````python
votos_por_jogador = [0] * 24  

print("Enquete: Quem foi o melhor jogador?\n")


total_votos = 0


while True:
    voto = int(input("Número do jogador (0=fim): "))
    
    if voto == 0:
        break  
    
    if voto < 0 or voto > 23:
        print("Informe um valor entre 1 e 23 ou 0 para sair!")
        continue  
    
    total_votos += 1
    votos_por_jogador[voto] += 1


print("\nResultado da votação:\n")
print(f"Foram computados {total_votos} votos.\n")
print("Jogador    Votos          %")
print("-" * 30)

maior_votos = 0
melhor_jogador = 0

for jogador, votos in enumerate(votos_por_jogador):
    if votos > 0:
        percentual = (votos / total_votos) * 100
        print(f"{jogador:<11} {votos:<15} {percentual:.1f}")
        
        if votos > maior_votos:
            maior_votos = votos
            melhor_jogador = jogador

print(f"\nO melhor jogador foi o número {melhor_jogador}, com {maior_votos} votos de um total de {total_votos}.")
````

9. Uma empresa de pesquisas precisa tabular os resultados da seguinte enquete feita a um grande quantidade de organizações:

```text
"Qual o melhor Sistema Operacional para uso em servidores?"

As possíveis respostas são:

1- Windows Server
2- Unix
3- Linux
4- Netware
5- Mac OS
6- Outro
```

Você foi contratado para desenvolver um programa que leia o resultado da enquete e informe ao final o resultado da mesma. O programa deverá ler os valores até ser informado o valor 0, que encerra a entrada dos dados. Não deverão ser aceitos valores além dos válidos para o programa (0 a 6). Os valores referentes a cada uma das opções devem ser armazenados em uma lista. Após os dados terem sido completamente informados, o programa deverá calcular o total de votos de cada um dos
concorrentes e informar o vencedor da enquete. 

A tela abaixo é um exemplo de execução do programa, apenas para fins ilustrativos. Os valores podem mudar a cada execução do programa.

```text
Sistema Operacional     Votos  
-------------------     -----  
Windows Server           1500  
Unix                     3500
Linux                    3000   
Netware                   500   
Mac OS                    150   
Outro                     150   
-------------------     -----
Total                    8800

O Sistema Operacional mais votado foi o Unix, com 3.500 votos.
```

````python
votos_por_sistema = [0] * 7  


sistemas_operacionais = [
    "Windows Server",
    "Unix",
    "Linux",
    "Netware",
    "Mac OS",
    "Outro"
]

print("Enquete: Qual o melhor Sistema Operacional para uso em servidores?\n")


while True:
    print("As possíveis respostas são:")
    for i, sistema in enumerate(sistemas_operacionais, start=1):
        print(f"{i}-{sistema}")
    print("0-Sair")
    
    voto = int(input("Informe o número correspondente ao Sistema Operacional (0 a 6): "))
    
    if voto == 0:
        break  
    
    if voto < 0 or voto > 6:
        print("Valor inválido. Por favor, digite um número de 0 a 6.")
        continue  
    
    votos_por_sistema[voto] += 1


print("\nResultado da votação:\n")
print("Sistema Operacional       Votos")
print("-" * 30)

total_votos = sum(votos_por_sistema)

for i, sistema in enumerate(sistemas_operacionais, start=1):
    print(f"{sistema:<26} {votos_por_sistema[i]}")
    
print("-" * 30)
print(f"Total                      {total_votos}\n")


mais_votado = max(range(1, 7), key=votos_por_sistema.__getitem__)

print(f"O Sistema Operacional mais votado foi o {sistemas_operacionais[mais_votado]}, com {votos_por_sistema[mais_votado]} votos.")
````

10. As Organizações Tabajara resolveram dar um abono aos seus colaboradores em reconhecimento ao bom resultado alcançado durante o ano que passou. Para isto contratou você para desenvolver a aplicação que servirá como uma projeção de quanto será gasto com o pagamento deste abono. Após reuniões envolvendo a diretoria executiva, a diretoria financeira e os representantes do sindicato laboral, chegou-se a seguinte forma de cálculo:

    - Cada funcionário receberá o equivalente a 20% do seu salário bruto de dezembro;
    - O piso do abono será de 100 reais, isto é, aqueles funcionários cujo salário for muito baixo, recebem este valor mínimo;

Neste momento, não se deve ter nenhuma preocupação com colaboradores com tempo menor de casa, descontos, impostos ou outras particularidades.

Seu programa deverá permitir a digitação do salário de um número indefinido (desconhecido) de salários. Um valor de salário igual a 0 (zero) encerra a digitação. Após a entrada de todos os dados o programa deverá calcular o valor do abono concedido a cada colaborador, de acordo com a regra definida acima. Ao final, o programa deverá apresentar:

- O salário de cada funcionário, juntamente com o valor do abono;
- O número total de funcionário processados;
- O valor total a ser gasto com o pagamento do abono;
- O número de funcionário que receberá o valor mínimo de 100 reais;
- O maior valor pago como abono; 

A tela abaixo é um exemplo de execução do programa, apenas para fins ilustrativos. Os valores podem mudar a cada execução do programa.

```text
Projeção de Gastos com Abono
============================

Salário: 1000
Salário: 300
Salário: 500
Salário: 100
Salário: 4500
Salário: 0

Salário    - Abono
R$ 1000.00 - R$  200.00
R$  300.00 - R$  100.00
R$  500.00 - R$  100.00
R$  100.00 - R$  100.00
R$ 4500.00 - R$  900.00

Foram processados 5 colaboradores
Total gasto com abonos: R$ 1400.00
Valor mínimo pago a 3 colaboradores
Maior valor de abono pago: R$ 900.00

````python
total_colaboradores = 0
total_gasto_abonos = 0
total_minimo_pago = 0
maior_abono_pago = 0

print("Projeção de Gastos com Abono")
print("============================\n")


while True:
    salario = float(input("Salário: "))
    
    if salario == 0:
        break  
    
    total_colaboradores += 1
    
    
    abono = salario * 0.2 if salario * 0.2 >= 100 else 100
    
    
    total_gasto_abonos += abono
    
    
    if abono == 100:
        total_minimo_pago += 1
    
    
    if abono > maior_abono_pago:
        maior_abono_pago = abono
    
    
    print(f"Salário: R$ {salario:.2f} - Abono: R$ {abono:.2f}")


print("\nForam processados", total_colaboradores, "colaboradores")
print("Total gasto com abonos: R$", total_gasto_abonos)
print("Valor mínimo pago a", total_minimo_pago, "colaboradores")
print("Maior valor de abono pago: R$", maior_abono_pago)
````

11. Faça um programa que carregue uma lista com os modelos de cinco carros (exemplo de modelos: FUSCA, GOL, VECTRA etc). Carregue uma outra lista com o consumo desses carros, isto é, quantos quilômetros cada um desses carros faz com um litro de combustível. Calcule e mostre:

    - O modelo do carro mais econômico;
    - Quantos litros de combustível cada um dos carros cadastrados consome para percorrer uma distância de 1000 quilômetros e quanto isto custará, considerando um que a gasolina custe 2,25 reais o litro.

Abaixo segue uma tela de exemplo. O disposição das informações deve ser o mais próxima possível ao exemplo. Os dados são fictícios e podem mudar a cada execução do programa.

```text
Comparativo de Consumo de Combustível

Veículo 1
Nome: fusca
Km por litro: 7
Veículo 2
Nome: gol
Km por litro: 10
Veículo 3
Nome: uno
Km por litro: 12.5
Veículo 4
Nome: Vectra
Km por litro: 9
Veículo 5
Nome: Peugeout
Km por litro: 14.5

Relatório Final
 1 - fusca           -    7.0 -  142.9 litros - R$ 321.43
 2 - gol             -   10.0 -  100.0 litros - R$ 225.00
 3 - uno             -   12.5 -   80.0 litros - R$ 180.00
 4 - vectra          -    9.0 -  111.1 litros - R$ 250.00
 5 - peugeout        -   14.5 -   69.0 litros - R$ 155.17
O menor consumo é do peugeout.
```

````python
modelos = ["FUSCA", "GOL", "UNO", "VECTRA", "PEUGEOUT"]
consumo_km_litro = [7, 10, 12.5, 9, 14.5]


litros_1000_km = [1000 / consumo for consumo in consumo_km_litro]
custo_1000_km = [litros * 2.25 for litros in litros_1000_km]


indice_carro_mais_economico = consumo_km_litro.index(max(consumo_km_litro))


print("Comparativo de Consumo de Combustível\n")

for i in range(len(modelos)):
    print(f"Veículo {i + 1}")
    print(f"Nome: {modelos[i]}")
    print(f"Km por litro: {consumo_km_litro[i]}\n")

print("Relatório Final")
for i in range(len(modelos)):
    print(f"{i + 1} - {modelos[i]:<15} - {consumo_km_litro[i]:>6.1f} - {litros_1000_km[i]:>7.1f} litros - R$ {custo_1000_km[i]:.2f}")

print(f"O menor consumo é do {modelos[indice_carro_mais_economico]}.")
````
